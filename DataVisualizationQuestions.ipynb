```python
#Create a bar chart showing the top 5 EV makes and models by count.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "Electric_Vehicle_Population_Data.csv"
df = pd.read_csv(file_path)

df['Make'] = df['Make'].astype(str).str.strip().str.title()
df['Model'] = df['Model'].astype(str).str.strip().str.title()

top_makes = df['Make'].value_counts().head(5)

top_models = df['Model'].value_counts().head(5)

plt.figure(figsize=(10, 4))
sns.barplot(x=top_makes.index, y=top_makes.values, palette='pastel')
plt.title("Top 5 EV Makes by Count")
plt.ylabel("Number of Vehicles")
plt.xlabel("EV Make")
plt.grid(axis='y')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
sns.barplot(x=top_models.index, y=top_models.values, palette='muted')
plt.title("Top 5 EV Models by Count")
plt.ylabel("Number of Vehicles")
plt.xlabel("EV Model")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#Use a heatmap or choropleth map to visualize EV distribution by county.
import pandas as pd
import plotly.express as px
import requests

file_path = "Electric_Vehicle_Population_Data.csv"
df = pd.read_csv(file_path)

df['County'] = df['County'].astype(str).str.strip().str.title()

county_counts = df['County'].value_counts().reset_index()
county_counts.columns = ['County', 'EV Count']

geojson_url = "https://opendata.arcgis.com/datasets/ba3b8d65e1914b82a6cddc5e28c65c3d_2.geojson"
geojson_data = requests.get(geojson_url).json()

county_counts['County'] = county_counts['County'].str.upper()

fig = px.choropleth(
    county_counts,
    geojson=geojson_data,
    locations='County',
    featureidkey='properties.NAME',
    color='EV Count',
    color_continuous_scale='YlGnBu',
    title='📍 EV Registrations by County (Washington State)',
    scope="usa",
)

fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})
fig.show()
```

```python
#Create a line graph showing the trend of EV adoption by model year.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "Electric_Vehicle_Population_Data.csv"
df = pd.read_csv(file_path)

df['Model Year'] = pd.to_numeric(df['Model Year'], errors='coerce')

df = df.dropna(subset=['Model Year'])

model_year_counts = df['Model Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x=model_year_counts.index, y=model_year_counts.values, marker='o', color='blue')

plt.title("📈 Trend of EV Adoption by Model Year", fontsize=14)
plt.xlabel("Model Year")
plt.ylabel("Number of EVs Registered")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Generate a scatter plot comparing electric range vs. base MSRP to see pricing trends.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "Electric_Vehicle_Population_Data.csv"
df = pd.read_csv(file_path)

df['Electric Range'] = pd.to_numeric(df['Electric Range'], errors='coerce')
df['Base MSRP'] = pd.to_numeric(df['Base MSRP'], errors='coerce')

df = df.dropna(subset=['Electric Range', 'Base MSRP'])
df = df[(df['Electric Range'] > 0) & (df['Base MSRP'] > 0)]

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Electric Range',
    y='Base MSRP',
    data=df,
    hue='Make',
    palette='tab10',
    alpha=0.7,
    edgecolor=None
)

plt.title("🔋 Electric Range vs 💰 Base MSRP", fontsize=14)
plt.xlabel("Electric Range (miles)")
plt.ylabel("Base MSRP (USD)")
plt.grid(True)
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Make')
plt.show()
```

```python
#Plot a pie chart showing the proportion of CAFV-eligible vs. non-eligible EVs.
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Electric_Vehicle_Population_Data.csv"
df = pd.read_csv(file_path)

caf_col = 'Clean Alternative Fuel Vehicle (CAFV) Eligibility'
df[caf_col] = df[caf_col].astype(str).str.strip().str.upper()

caf_counts = df[caf_col].value_counts()

plt.figure(figsize=(6, 6))
colors = ['#4CAF50', '#FF6F61', '#B0BEC5']  # Green, red, grey
plt.pie(
    caf_counts,
    labels=caf_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors[:len(caf_counts)]
)

plt.title("♻️ CAFV Eligibility of EVs", fontsize=14)
plt.axis('equal')  # Ensures pie is circular
plt.tight_layout()
plt.show()
```

```python
#Use a geospatial map to display EV registrations based on vehicle location.
import pandas as pd
import folium
from folium.plugins import MarkerCluster

file_path = "Electric_Vehicle_Population_Data.csv"
df = pd.read_csv(file_path)

def extract_coordinates(point):
    try:
        lon, lat = point.replace("POINT (", "").replace(")", "").split()
        return float(lat), float(lon)
    except:
        return None, None

df['Latitude'], df['Longitude'] = zip(*df['Vehicle Location'].map(extract_coordinates))

df = df.dropna(subset=['Latitude', 'Longitude'])
df = df[(df['Latitude'].between(-90, 90)) & (df['Longitude'].between(-180, 180))]

map_center = [47.5, -120.5]  # Approx center of WA
ev_map = folium.Map(location=map_center, zoom_start=7)

marker_cluster = MarkerCluster().add_to(ev_map)

for idx, row in df.iterrows():
    popup_text = f"""
    <b>Make:</b> {row['Make']}<br>
    <b>Model:</b> {row['Model']}<br>
    <b>Year:</b> {row['Model Year']}<br>
    <b>County:</b> {row['County']}
    """
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=popup_text,
        icon=folium.Icon(color="blue", icon="car", prefix="fa")
    ).add_to(marker_cluster)

ev_map
```
